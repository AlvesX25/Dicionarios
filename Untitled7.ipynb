```python
# criando um dicionario
pessoa = {"Nome": "Anderson", "Idade": 28 , "Estado": "Pernambuco", "Cidade": "Recife"}

# Exibindo o dicionario
print("Meu dicionáro de exemplo é:", pessoa)

# Acessando valores por chaves
print(pessoa["Nome"])
print(pessoa["Idade"])
print(pessoa["Cidade"])
print(pessoa["Estado"])

# Adicionando uma chave valor depois que o dicionário ja foi criado
pessoa["Profissão"] = "Engenheiro de Software"
print(pessoa)
```
