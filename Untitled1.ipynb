```python
import random

numero_secreto = random.randint(1, 10)
palpite = 0 # Começamos com 0 para o loop iniciar

print("--- JOGO DA ADIVINHAÇÃO ---")

# Enquanto o palpite for diferente (!=) do secreto, o jogo continua
while palpite != numero_secreto:
    palpite = int(input("Qual seu palpite (1 a 10)? "))

    if palpite == numero_secreto:
        print("🎉 Parabéns! Você acertou!")
    elif palpite > numero_secreto:
        print("Muito alto! Tente um número menor.")
    else:
        print("Muito baixo! Tente um número maior.")

print("Fim de jogo!")
```

```python
lista = []
num_compras = int(input("Quantos itens você deseja adicionar? "))

for i in range(num_compras):
    item = input(f"Digite o {i+1}º item: ")
    lista.append(item)

print(f"Sua lista de compras é: {lista}")

remove = input("Qual item deseja remover da lista? ")
if remove in lista:
    lista.remove(remove)
    print(f"'{remove}' foi removido. Nova lista: {lista}")
else:
    print(f"'{remove}' não foi encontrado na lista.")
```

```python
while True:
    senha = input("Digite sua senha: ")

    if len(senha) < 6:
        print("Senha muito curta. A senha deve ter no mínimo 6 caracteres.")
    else:
        confirmacao_senha = input("Digite novamente para confirmar a senha: ")

        if senha == confirmacao_senha:
            print("Senha cadastrada com sucesso!")
            break # Sai do loop se a senha for válida e confirmada
        else:
            print("As senhas não correspondem. Tente novamente.")
```

```python
import getpass # Importamos a biblioteca necessária

while True:
    # Em vez de input(), usamos getpass.getpass()
    senha = getpass.getpass("Crie sua senha: ")

    if len(senha) < 6:
        print("Erro: A senha precisa ter pelo menos 6 caracteres.")
    else:
        confirmacao = getpass.getpass("Confirme sua senha: ")

        if senha == confirmacao:
            print("Sucesso: Senha cadastrada!")
            break
        else:
            print("Erro: As senhas não coincidem. Recomece.")
```

```python
trabalho = True

while trabalho:
    print("Natam está trabalhando, Anderson está programando...")

    # O input precisa ficar DENTRO do while para perguntar a cada rodada
    usuario = input("Você já largou? Sim ou não: ").lower()

    # Forma correta de comparar a mesma variável com dois valores
    if usuario == "nao" or usuario == "não":
        trabalho = True
    else:
        trabalho = False

# O print de encerramento deve ficar FORA do while
print("Programa encerrado.")
```
